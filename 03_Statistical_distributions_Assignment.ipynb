```python
#Q1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

mu,std,size=5,3,10000
dist= np.random.normal(mu,std,size)
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
plt.hist(dist,bins=200)
plt.title('The distribution of Dist',color='darkred', size=15)
plt.xlabel('Numbers of Dist',color='darkblue', size=12)
plt.ylabel('Count of Numbers in the Dist',color='darkblue', size=12)


from scipy import stats
plt.subplot(1,2,2)
res = stats.probplot(dist, plot=plt)
plt.title("Q-Q Plot", color='darkred', size=15)
plt.xlabel("Theoretical Values", color='darkblue', size=12)
plt.ylabel(" Data",color='darkblue', size=12)
plt.show()

print('The Data is normally distributed')
```

```python
#Q2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas_datareader import data as pdr
import yfinance as yf

data_google = pdr.get_data_yahoo("GOOG", start="2015-01-01", end="2017-01-01")
data_google.head()
stock_return=[]
for value in range(1,len(data_google.Close)-1):
    a=(data_google.Close[value]-data_google.Close[value-1])/data_google.Close[value-1]
    stock_return.append(a)

plt.figure(figsize=(15,4),dpi=300)
plt.subplot(1,2,1)
plt.hist(stock_return)


from scipy import stats
plt.subplot(1,2,2)
res = stats.probplot(stock_return, plot=plt)
plt.title("Q-Q Plot", color='darkred', size=15)
plt.xlabel("Theoretical Values", color='darkblue', size=12)
plt.ylabel(" Data",color='darkblue', size=12)
plt.show()

print("According to Histogram and Q-Q Plot, it can be said that the Google's Data between 2015-2017  is normally distributed")
```
