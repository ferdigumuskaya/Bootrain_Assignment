In this assignment, you will make some plots about Corona. First, download the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data).

- First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. 

- Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

- Set titles, X and Y labels, change the font, font size, and font color for each subplot.

- Change color, width, and type of lines.

- Put a legend to bottom right.

- Change the color and rotation of the X-axis ticks.

- And set a title for the figure (not for the plots).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#1st bullet
corona = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv',low_memory=False, date_parser = ['date'])
n_corona=corona[(corona['location'].isin(['Spain','France','Germany','Italy']))& (corona['date']>'2020-04-22')]
n_corona

```

```python
#2nd bullet
plt.figure(figsize=(24,6))

plt.subplot(1,2,1)
plt.plot(n_corona[n_corona['location']=='Spain']['date'],n_corona[n_corona['location']=='Spain']['total_cases'])
plt.plot(n_corona[n_corona['location']=='France']['date'],n_corona[n_corona['location']=='France']['total_cases'])
plt.plot(n_corona[n_corona['location']=='Germany']['date'],n_corona[n_corona['location']=='Germany']['total_cases'])
plt.plot(n_corona[n_corona['location']=='Italy']['date'],n_corona[n_corona['location']=='Italy']['total_cases'])


plt.subplot(1,2,2)
plt.plot(n_corona[n_corona['location']=='Spain']['date'],n_corona[n_corona['location']=='Spain']['total_deaths'])
plt.plot(n_corona[n_corona['location']=='France']['date'],n_corona[n_corona['location']=='France']['total_deaths'])
plt.plot(n_corona[n_corona['location']=='Germany']['date'],n_corona[n_corona['location']=='Germany']['total_deaths'])
plt.plot(n_corona[n_corona['location']=='Italy']['date'],n_corona[n_corona['location']=='Italy']['total_deaths'])

plt.show()
```

```python
#3rd bullet
plt.figure(figsize=(24,6))
title_style = {'family': 'Arial', 'color': 'darkred', 'size': 20 }
axis_style  = {'family': 'Arial', 'color': 'darkblue', 'size': 15 }

plt.subplot(1,2,1)
plt.title('The Total Cases',fontdict=title_style)
plt.xlabel('The Last 100 days',fontdict=axis_style)
plt.ylabel('The number of Total Cases',fontdict=axis_style)
plt.plot(n_corona[n_corona['location']=='Spain']['date'],n_corona[n_corona['location']=='Spain']['total_cases'])
plt.plot(n_corona[n_corona['location']=='France']['date'],n_corona[n_corona['location']=='France']['total_cases'])
plt.plot(n_corona[n_corona['location']=='Germany']['date'],n_corona[n_corona['location']=='Germany']['total_cases'])
plt.plot(n_corona[n_corona['location']=='Italy']['date'],n_corona[n_corona['location']=='Italy']['total_cases'])


plt.subplot(1,2,2)
plt.title('The Total Deaths',fontdict=title_style)
plt.xlabel('The Last 100 days',fontdict=axis_style)
plt.ylabel('The number of Total Deaths',fontdict=axis_style)
plt.plot(n_corona[n_corona['location']=='Spain']['date'],n_corona[n_corona['location']=='Spain']['total_deaths'])
plt.plot(n_corona[n_corona['location']=='France']['date'],n_corona[n_corona['location']=='France']['total_deaths'])
plt.plot(n_corona[n_corona['location']=='Germany']['date'],n_corona[n_corona['location']=='Germany']['total_deaths'])
plt.plot(n_corona[n_corona['location']=='Italy']['date'],n_corona[n_corona['location']=='Italy']['total_deaths'])

plt.show()
```

```python
#4th bullet
plt.figure(figsize=(24,6))
title_style = {'family': 'Arial', 'color': 'darkred', 'size': 20 }
axis_style  = {'family': 'Arial', 'color': 'darkblue', 'size': 15 }

plt.subplot(1,2,1)
plt.title('The Total Cases',fontdict=title_style)
plt.xlabel('The Last 100 days',fontdict=axis_style)
plt.ylabel('The number of Total Cases',fontdict=axis_style)
plt.plot(n_corona[n_corona['location']=='Spain']['date'],n_corona[n_corona['location']=='Spain']['total_cases'],c='red',lw=1,ls="dashed")
plt.plot(n_corona[n_corona['location']=='France']['date'],n_corona[n_corona['location']=='France']['total_cases'],c='green',lw=1,ls="dashed")
plt.plot(n_corona[n_corona['location']=='Germany']['date'],n_corona[n_corona['location']=='Germany']['total_cases'],c='blue',lw=1,ls="dashed")
plt.plot(n_corona[n_corona['location']=='Italy']['date'],n_corona[n_corona['location']=='Italy']['total_cases'],c='black',lw=1,ls="dashed")


plt.subplot(1,2,2)
plt.title('The Total Deaths',fontdict=title_style)
plt.xlabel('The Last 100 days',fontdict=axis_style)
plt.ylabel('The number of Total Deaths',fontdict=axis_style)
plt.plot(n_corona[n_corona['location']=='Spain']['date'],n_corona[n_corona['location']=='Spain']['total_deaths'],c='red',lw=1,ls="dashed")
plt.plot(n_corona[n_corona['location']=='France']['date'],n_corona[n_corona['location']=='France']['total_deaths'],c='green',lw=1,ls="dashed")
plt.plot(n_corona[n_corona['location']=='Germany']['date'],n_corona[n_corona['location']=='Germany']['total_deaths'],c='blue',lw=1,ls="dashed")
plt.plot(n_corona[n_corona['location']=='Italy']['date'],n_corona[n_corona['location']=='Italy']['total_deaths'],c='black',lw=1,ls="dashed")

plt.show()
```
