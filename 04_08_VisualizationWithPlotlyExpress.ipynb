In this assignment you will continue to make some plots on the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data). For plotting you will use Plotly Express. 

```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv', parse_dates=["date"], low_memory=False)
df.head()
```

1. Plot a line plot with plotly express for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
df1=df[df.date >'2020-04-01']

fig=px.line(df1, x= df1.date, y=df1.total_cases,color=df1.location,title='Total Deaths of four given countries',width = 700, height = 400)
fig.show()
```

2. Plot a bar plot with plotly express for average death number that compares continents.

```python
df2=df.groupby(by='continent')['total_deaths'].mean()

px.bar(df2,x= df2.index,y=df2,width = 700, height = 400)
```

3. Plot a histogram for daily deaths for any country you choose. 

```python
df3= df[df.location =='Germany']
fig=px.histogram(df3, x=df3.date, y=df3.new_deaths)
fig.show()
```
