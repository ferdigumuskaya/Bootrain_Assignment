````python
import pandas as pd

feature = pd.DataFrame({'id': [1001,1001,1002,1002,1003,1004,1005,1006,1007,1007,1008,1008],
                        'age': [63,37,41,56,57,57,56,44,52,57,54,48],
                        'sex': ['Male','Male','Female','Male','Female','Male','Female','Male','Male','Male','Male','Female'],
                        'beats_per_sec': [145,130,130,120,120,140,140,120,172,150,140,150],
                        'cholesterol': [233,250,204,236,354,192,294,263,199,168,239,264]
                       })
target = pd.DataFrame ({ 'id': [1001,1002,1004,1005,1007,1008],
                        'target': [0,1,0,1,0,1]
                       })

display(feature)
display(target)

#Inner Join
inner_merge = pd.merge(feature,target, on='id', how='inner')
display(inner_merge)
#Outer Join
outer_merge = pd.merge(feature,target, on='id', how='outer')
display(outer_merge)
#Left Join
left_join = pd.merge(feature,target, on='id', how='left')
display(left_join)
#Right Join
right_join = pd.merge(feature,target, on='id', how='right')
display(right_join)
````
